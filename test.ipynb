```python
from matplotlib import pyplot as plt
import numpy as np

# start position
x_start = 0. 
y_start = 1. 

# list with all positions
all_x = [x_start]
all_y = [y_start]

# show graph
plt.figure() 
plt.plot(all_x, all_y, '-o')
  
```

```python
# nicer graph 
def show_graph(all_x, all_y): 
    plt.figure() 
    plt.plot(all_x, all_y, '-o')
    # plot floor 
    plt.plot([0., 10.],[0., 0.], color='black')
    plt.axis([0., 10., 0., 6.])

show_graph(all_x, all_y) 
```

```python
# initial velocity
v_x_start = 3. 
v_y_start = 3. 

# move one step (constant velocity): s = v*t 
delta_t = 0.1
x_end = x_start + delta_t*v_x_start 
y_end = y_start + delta_t*v_y_start 

# add final position to list of all positions 
all_x.append(x_end)
all_y.append(y_end)

# show graph
show_graph(all_x, all_y)
```
